```python
class Solution:
    def insert(self, intervals: List[List[int]], newInterval: List[int]) -> List[List[int]]:
        """
        - First find where to place the interval based on the newInterval[0]
        i.e. compare wheter intervals[0] <= newInterval[0] <= intervals[1]

        - If you find such a spot, 

        - if you dont find a in-between position, just append and sort based on the initial number
        """
        res = []

        for i in range(len(intervals)):
            if newInterval[0] > intervals[i][1]:
                res.append(intervals[i])
            elif newInterval[1] < intervals[i][0]:
                res.append(newInterval)
                #no need to check for anything else if this condition holds
                return res + intervals[i:]

            else:
                newInterval = [min(newInterval[0],intervals[i][0] ), max(newInterval[1],intervals[i][1] ) ]
        res.append(newInterval)

        return res
```

973. K Closest Points to Origin: https://leetcode.com/problems/k-closest-points-to-origin/description/

```python
class Solution:
    def kClosest(self, points: List[List[int]], k: int) -> List[List[int]]:
        """
        - store the answer in an array that is of k size
        - find the distance to the origin from each point 
        - make a hashmap for each {point: distance} 
        - take the first k lowest distances and their points 
        (we could add a third parameter in the list and sort it based on it)
        """
        points_distance = []
        for point in points:
            value =  point[0]**2 + point[1]**2
            points_distance.append( [point[0], point[1], value ] )

        points_distance.sort(key = lambda distance: distance[2] )

        # print(points_distance[:k])
        new_points = []
        for point in points_distance[:k]:

            new_points.append([point[0], point[1]])
        # print(new_points)
        return new_points
        
```

Clone Graph: https://leetcode.com/problems/clone-graph/description/

```python
"""
# Definition for a Node.
class Node:
    def __init__(self, val = 0, neighbors = None):
        self.val = val
        self.neighbors = neighbors if neighbors is not None else []
"""

class Solution:

    def __init__(self):
        self.visited = {}

    def cloneGraph(self, node: 'Node') -> 'Node':
        # print(node)
        if not node:
            return node
        if node in self.visited:
            return self.visited[node]

        
        cloned_node = Node(node.val, [])
        self.visited[node] = cloned_node

        if node.neighbors:
            cloned_node.neighbors = [ self.cloneGraph(n) for n in node.neighbors]
        
        return cloned_node
```

150. Evaluate Reverse Polish Notation: https://leetcode.com/problems/evaluate-reverse-polish-notation/description/

```python
import operator
class Solution:
    def evalRPN(self, tokens: List[str]) -> int:
        """
        Use 
        """
        if len(tokens) == 1:
            return int(tokens[0])

        ops = { "+": operator.add, "-": operator.sub, "*":operator.mul, "/": operator.truediv }
        # hashmap = {'+':+ , '-':- '*':*, '/':/}
        stack = []
        for i in range(len(tokens)):
            
            if tokens[i] in ['+', '-', '*', '/']:
                # print('yes')
                number1 = int(stack.pop())
                # print(stack)
                number2 = int(stack.pop())
                
                # sign = stack.pop(i)
                result = ops[tokens[i]](number2, number1)
                # print(result)
                stack.append(result)
            else:
                stack.append(int(tokens[i]))
            # print(stack)
        return int(stack[0])
```
